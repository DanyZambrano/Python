Mergear un dataframe integra filas y columnas de diveros dataframes que existan con informacion en comun.

```python
import pandas as pd
```

```python
ruta = '/Users/danyzambrano/Desktop/pandas/example1.xlsx'
dataframe1 = pd.read_excel(ruta)
dataframe1
```

```python
ruta2 = '/Users/danyzambrano/Desktop/pandas/example2.xlsx'
dataframe2 = pd.read_excel(ruta2)
dataframe2
```

outer: crear columnas de las que no se pudieron emparejar

```python
nuevo_dataframe = pd.merge(dataframe1, dataframe2, on="First Name", how="outer")
nuevo_dataframe
```

left: descarta las que no se pudieron emparejar

```python
nuevo_dataframe = pd.merge(dataframe1, dataframe2, on="First Name", how="left")
nuevo_dataframe
```

right: descarta la filas del segundo dataframe que no se pudieron emparejar

```python
nuevo_dataframe = pd.merge(dataframe1, dataframe2, on="First Name", how="right")
nuevo_dataframe
```

```python
nuevo_dataframe = pd.merge(dataframe1, dataframe2, on=["First Name", "Last Name", "Gender", "Country", "Age", "Id"], how="outer")
nuevo_dataframe
```
