Como es la estructura para crear un dataframe a nivel de codigo?

DataFrame(data=diccionario, index=filas, columns=columnas, dtype=tipos).

La mejor forma de crear un dataframe en codigo es con un diccionario como data de entrada.

```python
import pandas as pd
```

```python
datos = {
'nombre':['Dany', 'Alexander', 'Santiago', 'Samuel', 'Sebastian'],
'correo':['dany@micorreo.com', 'alexander@micorreo.com', 'santiago@micorreo.com', 'samuel@micorreo.com', 'sebastian@micorreo.com'],
'deporte':['Domino', 'Basket', 'Soccer', 'Baseball', 'Soccer'],
'trofeos': [23, 21, 34, 34, 87],
'campeonatos': [False, True, True, True, True],
 }
```

```python
df = pd.DataFrame(datos)
df
```

Dataframe con indices customizados

```python
datos = {
'nombre':['Dany', 'Alexander', 'Santiago', 'Samuel', 'Sebastian'],
'correo':['dany@micorreo.com', 'alexander@micorreo.com', 'santiago@micorreo.com', 'samuel@micorreo.com', 'sebastian@micorreo.com'],
'deporte':['Domino', 'Basket', 'Soccer', 'Baseball', 'Soccer'],
'trofeos': [23, 21, 34, 34, 87],
'campeonatos': [False, True, True, True, True],
 }
```

```python
df = pd.DataFrame(datos, index=['usuario_1','usuario_2','usuario_3','usuario_4','usuario_5'])
df
```

```python
df = pd.DataFrame(datos, index=['dan','alex','san','sam','sebas'])
df
```

Como Obtener Valores de una Columna?

```python
df['trofeos']
```

```python
df.nombre
```
