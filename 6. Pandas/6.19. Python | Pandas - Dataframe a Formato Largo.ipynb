```python
ruta = '/Users/danyzambrano/Desktop/pandas/example_excel.xlsx'
data = pd.read_excel(ruta).head(15)
data
```

Convertir un Dataframe a formato largo

```python
df = data.melt(id_vars=['First Name'], var_name='Parametro', value_name='Valor')
df
```

Convertir un DataFrame a formato Ancho

```python
df.pivot(index='First Name', columns='Parametro', values='Valor')
```
